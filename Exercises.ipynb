### Given a number, replace all 0's into 1's.

```python
value = input()
value_str = str(value)
char_arr = list(value_str)
```

```python
# Turns out that which() is an R function.

#indices = which(value) == "0"
#value[indices] = "1"
#return(value)
```

```python
for i = range(0, len(char_arr)):
    if char_arr[i] == "0":
        char_arr[i] = "1"
return value
```

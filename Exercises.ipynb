### Reverse an array

```python
v = [1, 5, 3, 4, 7, 2]
```

```python
end = len(v) -1

#This even variable ensures that the code is ran an even number of times, as the loop to reverse the array is based on the length of the
#list, to prevent reversing twice (hence appearing that the list is not switched).
even = end/2
```

```python
#Constant to subtract the length of the list to be used as an index.
C_end = len(v)-1
```

```python
while(end > even):
    #Save the last value of the list w as another variable
    v2 = v[end]

    v[end]=v[C_end - end]
    
    v[C_end - end]=v2
    print(v)
    end-=1
```

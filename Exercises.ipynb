```python
value = input()
#The values are converted to strings as to concatenate integers, value must be converted to a string.
value_str = str(value)
char_arr = list(value_str)
```

```python
# Turns out that which() is an R function.

#indices = which(value) == "0"
#value[indices] = "1"
#return(value)
```

```python
for i in range(0, len(char_arr)):
    if char_arr[i] == "0":
        char_arr[i] = "1"
```

```python
#The char_arr prints out the array of strings
#char_arr
value_1 = str()
for i in char_arr:
    value_1 = (value_1 + i)

print(int(value_1))
```
